```python
def q1():

    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.ticker as mtick
    import matplotlib.patches as mpatches
    from matplotlib.lines import Line2D
    import matplotlib
    
    df= pd.read_excel("iris_pasaport_applications.xlsx")
    x_axis=np.array(df["Year"])
    first_time=np.array(df["FirstTime"])
    renewal=np.array(df["Renewal"])
    renewal[0]+first_time[0]

    fig, ax = plt.subplots(dpi=300)

    plt.box(False)  # remove 

    fig.patch.set_facecolor('#FEF2E6')  # set figure background color
    
    fmt = "{x:,.0f}"
    tick = mtick.StrMethodFormatter(fmt)
    ax.yaxis.set_major_formatter(tick) 
    ax.yaxis.tick_right()  # get ticks right
    ax.grid(axis="y",zorder=0, linewidth=0.2) # get grid for y axis
    ax.tick_params(axis ='both', which ='both', length = 0) # remove tick markers
    #ax.grid(zorder=0)

    width = 0.8     
    p1 = ax.bar(x_axis[0], renewal[0], width, color='#D7CDC3', bottom=0, zorder=2)
    p5 = ax.bar(x_axis[1:5], renewal[1:5], width, color='#D7CDC3', bottom=0, zorder=2)
    p2 = ax.bar(x_axis[0], first_time[0], width, color='#FB78A4', zorder=3, bottom=renewal[0], edgecolor='black', linewidth=1.2) # outer line
    p3 = ax.bar(x_axis[4], first_time[4], width, color='#FB78A4', zorder=3, bottom=renewal[4], edgecolor='black', linewidth=1.2) # outer line
    p4 = ax.bar(x_axis[1:4], first_time[1:4], width, color='#FB78A4', zorder=3, bottom=renewal[1:4])

    ax.bar_label(p2,labels=["first time"], label_type='center', fontsize=8)
    ax.bar_label(p1,labels=["renewal"], label_type='center', fontsize=8)
    
    ax.annotate('There were 7,7372 first time\napplications in the year before\nthe referandum..', fontsize=8, xy=(2015,34000), xytext=(0, 75), xycoords="data",
            ha= "center",textcoords='offset points', arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=0.3",color="gray",linewidth=2));
    
    ax.annotate('.. in 2019, there\nwere over 54,000', fontsize=8, xy=(2018.6,70000), xytext=(-140, 35), xycoords="data",
            textcoords='offset points', arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=-0.3",color="gray",linewidth=2));
    
    ax.annotate("Applications from Great Britain-born applicants:",  xy=(40, 270),  xycoords='figure points',
                xytext=(0,0), textcoords='offset points',fontsize=8, color='gray')
    
    ax.annotate("Irish passport applications:",  xy=(40, 285),  xycoords='figure points',
                xytext=(0, 0),  fontsize=8, textcoords='offset points', color='black',weight="bold")

    ax.annotate("a Brexit-fuelled surge",  xy=(175, 300),  xycoords='figure points',
                xytext=(0, 0), fontsize=8, textcoords='offset points', color='black')
    
    ax.annotate("a Brexit-fuelled surge",  xy=(175, 285),  xycoords='figure points',
                xytext=(0, 0), fontsize=8, textcoords='offset points', color='black')
  
    return 
```

```python
def q2():
    
    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np
    import matplotlib.ticker as mtick
    import matplotlib.patches as mpatches
    from matplotlib.lines import Line2D
    import matplotlib

    country_index=["Chile","India","Canada","France","Italy","US","UK","Israel","Germany"]
    mobility=[-6,-2,-4,-2,-2,-1,3.5,3,2]
    case=[10,50,30,15,-10,5,-10,-30,20]
    colorlist=["navy","cyan","dodgerblue","cyan","cyan","dodgerblue","navy","navy","cyan"]

    fig, ax = plt.subplots(dpi=300)

    plt.box(False)  # remove 

    z=plt.scatter(mobility,case, c=colorlist, alpha=0.5, s=100,linestyle="None")

    ax.set_ylabel('Weekly change in new cases(%)',fontsize=9)
    ax.set_xlabel('Weekly change in mobility(%)', fontsize=9)

    ax.set_xticks(np.arange(-10, 11, 5))
    ax.set_yticks(np.arange(-60, 61, 20))

    ax.grid(axis="both", linewidth=0.3) # get grid for axes

    ax.axhline(y=0, color='0.5',lw=1.5)
    ax.axvline(x=0, color='0.5',lw=1.5)


    fig.patch.set_facecolor('#FEF2E6')  # set figure background color

    for i, txt in enumerate(country_index):
        ax.annotate(txt, (mobility[i], case[i]),fontsize=5.5)

    fig.patch.set_facecolor('#FEF2E6')
    ax.tick_params(axis ='both', which ='both', length = 0) # remove tick markers

    rect=mpatches.Rectangle((0,0),10,-60, fill = True,color = "lightgreen", alpha = 0.2, linewidth = 2)

    plt.gca().add_patch(rect)

    ax.annotate("Target zone",  xy=(315, 71),  xycoords='figure points',
            xytext=(0, 0), textcoords='offset points',fontsize=11, color='darkgreen')

    ax.annotate("Countries are seeking to fully reopen while keeping the cases decline",  xy=(0, 290),  xycoords='figure points',
                    xytext=(0, 0), textcoords='offset points', fontsize=9, color='green')

    ax.annotate("Change in mobility levels vs change in new cases (7-day averages)",  xy=(0, 280),  xycoords='figure points',
                    xytext=(0, 0), textcoords='offset points', fontsize=7, color='black')

    ax.annotate("Share of population with at least\none vaccine dose(%)",  xy=(260, 250),  xycoords='figure points',
                    xytext=(0, 0), textcoords='offset points',fontsize=8, color='black')

    legend_elements = [Line2D([0], [0], marker='o', color='w', label='   0-20',markerfacecolor='cyan', markersize=9),
                        Line2D([0], [0], marker='o', color='w', label='   20-40',markerfacecolor='navy', markersize=9),
                       Line2D([0], [0], marker='o', color='w', label='   40+',markerfacecolor='dodgerblue', markersize=9)]

    matplotlib.rcParams['legend.handlelength'] = 0
    matplotlib.rcParams['legend.numpoints'] = 1

    ax.legend(handles=legend_elements, loc=(0.76,0.83), framealpha=0, fontsize=7)
    
    return()
```

```python
q2()
```
